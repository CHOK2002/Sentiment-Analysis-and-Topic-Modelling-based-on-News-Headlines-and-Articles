Renamed columns in the cnn_news_articles_2023 dataset

```python
# load the CSV file
df = pd.read_csv('cnn_news_articles_2023.csv')

# rename the columns
df_renamed = df.rename(columns={
    'Title': 'Headline',
    'Body': 'Article text',
    'Theme': 'Category',
    'Link': 'Url'
})

# save the updated DataFrame back to a new CSV file
df_renamed.to_csv('cnn_news_articles_2023.csv', index=False)
```

Comparison (cnn_news_articles_2011_2022 and cnn_news_articles_2023)

```python
import datacompy

# load the datasets
cnn_news_articles_1 = pd.read_csv("cnn_news_articles_2011_2022.csv")
cnn_news_articles_2 = pd.read_csv("cnn_news_articles_2023.csv")

# initialize the DataCompare object
compare = datacompy.Compare(
    cnn_news_articles_1,
    cnn_news_articles_2,
    join_columns='Article text',  
    abs_tol=0,          
    rel_tol=0,          
    df1_name='cnn_news_articles_2011_2022',  
    df2_name='cnn_news_articles_2023'  
)

# run the comparison
comparison_results = compare.report()

# print the comparison results
print(comparison_results)
```
